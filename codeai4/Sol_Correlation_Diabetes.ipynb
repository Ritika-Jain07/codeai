```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/Professional50coder/KJSIT-AILabs/main/Project_4/diabetes_binary_5050split_health_indicators_BRFSS2015.csv"
df = pd.read_csv(url)

# Explore the dataset
df.head()
```

```python
df.info()
```

## Descriptive analysis

```python
df.describe()
```

```python
columns_to_drop = ['Income', 'Education','DiffWalk', 'PhysHlth','MentHlth', 'GenHlth','NoDocbcCost', 'AnyHealthcare','Fruits','Veggies']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
corr_matrix = df.corr()
fig , ax = plt.subplots(figsize=(10,8))
ax = sns.heatmap(corr_matrix,
                annot=True,
                linewidths=0.8,
                fmt='.2f')
```
